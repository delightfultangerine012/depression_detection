```python
base_directory = "/content/data/final-eriskt2-dataset-with-ground-truth"

data_directory = base_directory + "/all_combined"
label_file = base_directory + "/shuffled_ground_truth_labels.txt"
```

```python
from datetime import datetime
from dateutil.parser import parse
import re
import pandas as pd

import json
from sklearn.model_selection import train_test_split

from bs4 import BeautifulSoup
```

```python
users = []
labels = []

with open(label_file, "r") as f:
    for line in f:
        u, l = line.strip().split()
        users.append(u)
        labels.append(int(l))

df = pd.DataFrame({"user": users, "label": labels})

df["file"] = df["user"].apply(lambda u: os.path.join(data_directory, f"{u}.json"))

df = df[df["file"].apply(os.path.exists)].reset_index(drop=True)

print("dataset size:", len(df))
print(df["label"].value_counts())
```

```python
train_df, test_df = train_test_split(
    df,
    test_size=0.5,
    random_state=42,
    stratify=df["label"]
)

print("train:", len(train_df))
print("test:", len(test_df))
```

```python
print("train data balance:")
print(train_df["label"].value_counts())

print("\ntest data balance:")
print(test_df["label"].value_counts())
```

```python
def count_posts_comments(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    posts = len(data)
    comments = sum(len(post.get("comments", [])) for post in data)

    return posts, comments

train_df = train_df.copy()
test_df = test_df.copy()

train_df[["posts", "comments"]] = train_df["file"].apply(
    lambda x: pd.Series(count_posts_comments(x))
)

test_df[["posts", "comments"]] = test_df["file"].apply(
    lambda x: pd.Series(count_posts_comments(x))
)
```

```python
def summarise(df):

    summary = (
        df.groupby("label")
          .agg(
              Users=("user", "count"),
              Posts=("posts", "sum"),
              Comments=("comments", "sum")
          )
    )

    summary.index = summary.index.map({
        0: "control",
        1: "positive"
    })

    return summary

train_summary = summarise(train_df)
test_summary = summarise(test_df)

print("Training")
print(train_summary)

print("\nTest")
print(test_summary)
```

```python
def load_user(file_path):
    with open(file_path, "r") as f:
        return json.load(f)
```

```python
def flatten_sequence(seq):
    texts = []

    for thread in seq:


        submission = thread.get("submission", {})

        title = submission.get("title", "")
        body = submission.get("body", "")

        if title:
            texts.append(title)

        if body:
            texts.append(body)


        comments = thread.get("comments", [])

        for comment in comments:
            text = comment.get("body", "")

            if text:
                texts.append(text)

    return " ".join(texts)
```

```python
Maximum_Characters = 4000
```

```python
train_texts = []
train_labels = []

for _, row in train_df.iterrows():
    seq = load_user(row["file"])
    text = flatten_sequence(seq)[-Maximum_Characters:]

    train_texts.append(text)
    train_labels.append(row["label"])
```

```python
test_texts = []
test_labels = []

for _, row in test_df.iterrows():
    seq = load_user(row["file"])
    text = flatten_sequence(seq)[-Maximum_Characters:]

    test_texts.append(text)
    test_labels.append(row["label"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

vectorizer = TfidfVectorizer(
    max_features=40000,
    ngram_range=(1,2),
    min_df=2,
    stop_words="english"

)
```

```python
Xtrain = vectorizer.fit_transform(train_texts)
Xtest = vectorizer.transform(test_texts)
```

```python
model = LogisticRegression(
    max_iter=1000,
    class_weight="balanced"
)

model.fit(Xtrain, train_labels)
```

```python
preds = model.predict(Xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_labels, preds))
```

```python
from datasets import Dataset
train_dataset = Dataset.from_dict({
    "text": train_texts,
    "label": train_labels
})

test_dataset = Dataset.from_dict({
    "text": test_texts,
    "label": test_labels
})
```

```python
!pip install -U transformers
from transformers import AutoTokenizer

model_name_is = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name_is)
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=512
    )

train_dataset = train_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)
```

```python
train_dataset = train_dataset.remove_columns(["text"])
test_dataset = test_dataset.remove_columns(["text"])

train_dataset.set_format("torch")
test_dataset.set_format("torch")
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

```python
training_args = TrainingArguments(
    output_dir="./BERT_result",
    eval_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset

)
```
